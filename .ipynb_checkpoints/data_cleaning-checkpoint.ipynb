```python
import json
import pickle
```

```python
filename = "./data/data.json"
with open(filename, 'r') as f:
    datastore = json.load(f)
```

```python
data = datastore['result']['records']
data = [datum for datum in data if datum['Topics']!='']
print(len(data))
topics = dict()
for datum in data:
    datum['topics'] = []
    for key in datum['Topics'].split('|'):
        if key in topics:
            topics[key] += 1
        else:
            topics[key] = 1
        datum['topics'].append(key)
    #Remove original unprocessed "Topics"
    datum.pop('Topics', None)
```

```python
with open('topics.pickle', 'wb') as f:
    pickle.dump(topics, f)
```
